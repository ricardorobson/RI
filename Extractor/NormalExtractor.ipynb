# Extrator de informações

```python
import pandas as pd
import os
from bs4 import BeautifulSoup
import re
```

### Criando lista de páginas do dominio Psicologosdobrasil

```python
#https://www.psicologiaviva.com.br/rejapsi não funciona
#..\\Pages\\Positives\\psicologiaviva\\0.html

#https://www.psicologosdobrasil.com.br/detalhespsicologo.php?nome=Zara%20Bruno&p=98
#funciona

soups = []
files = len(os.listdir(os.path.abspath('..\\Pages\\Positives\\psicologosdobrasil\\')))
print(files)
for page in range(files):
    soups.append(BeautifulSoup(
        open(os.path.abspath('..\\Pages\\Positives\\psicologosdobrasil\\'+str(page)+'.html'), encoding="utf-8"), 'html.parser'))
soups[0].head
```

### Extraindo dados das páginas do Psicologosdobrasil

```python
names = []
formations = []
locations = []
descriptions = []
phones = []
for soup in soups:
    names.append(soup.find('h2').text)
    formations.append(soup.find('div', {'class': 'heading-block center'}).find('span').text)
    locations.append(soup.find('i', {'class': 'icon-map-marker2'}).parent.text)
    descriptions.append(soup.find('div', {'class': 'col_two_third'}).text)
    phones.append(soup.find('i', {'class': 'icon-phone'}).parent.text)

d = { 'nome': names, 'formação': formations, 'localização': locations, 'telefone':phones, 'descrição': descriptions }
df = pd.DataFrame(data=d)
df = df[['nome', 'formação', 'localização', 'telefone', 'descrição']]
df
```

### Extraindo paginas do domínio Doctoralia

```python
soups = []
files = len(os.listdir(os.path.abspath('..\\Pages\\Positives\\doctoralia\\')))
print(files)
for page in range(files):
    #throwing UnicodeDecodeErro 'charmap' codec can't decode byte 0x81
    soups.append(BeautifulSoup(
        open(os.path.abspath('..\\Pages\\Positives\\doctoralia\\'+str(page)+'.html'), errors="ignore", encoding="utf-8"), 'html.parser'))
soups[0].head
```

### Extraindo dados das páginas do Doctoralia

```python
names = []
specialities = []
crps = []
formations = []
locations = []
phones = []
for soup in soups:
    names.append(soup.find('h1').text)
    specialities.append(soup.find('div', {'id': 'doctorSpecialities'}).find('p').text)
    crp = soup.find('p', {'class': 'regnum'}).text[40:]
    crp = re.sub('[A-z]| |:|-', '', crp)
    crps.append(crp)
    education = '; '.join(str(x.text)
                         for x in soup.find('section', {'class': 'education'}).find_all('li'))
    formations.append(education)
    location = soup.find("span", {"class": "location"}).text.strip()
    if('\n' in location):
        location = location[:location.index('\n')]
    locations.append(location)
    #descriptions.append(soup.find('div', {'class': 'col_two_third'}).text)
    #phones.append(soup.find("i", {"class": "icon-phone"}).parent.text)

d = { 'nome': names, 'especialidade': specialities, 'CRP': crps, 'formação': formations, 'localização': locations }
     #, 'formação': formations, 'localização': locations, 'telefone':phones, 'descrição': descriptions }
df = pd.DataFrame(data=d)
df = df[['nome', 'CRP', 'localização', 'especialidade', 'formação']]
df
```

### Extraindo paginas do domínio Mundopsicologos

```python
soups = []
files = len(os.listdir(os.path.abspath('..\\Pages\\Positives\\\mundopsicologos\\')))
print(files)
for page in range(files):
    #throwing UnicodeDecodeErro 'charmap' codec can't decode byte 0x81
    soups.append(BeautifulSoup(
        open(os.path.abspath('..\\Pages\\Positives\\\mundopsicologos\\'+str(page)+'.html'), errors="ignore", encoding="utf-8"), 'html.parser'))
soups[0].head
```

### Extraindo dados das páginas do Mundopsicologos

```python
names = []
responsibles = []
especialities = []
descriptions = []
services = []
adresses = []

for soup in soups:
    names.append(soup.find('h1').text)
    table = soup.find_all('ul')
    responsibles.append(table[1].text.strip())
    especialities.append(table[2].text.strip())
    descriptions.append(soup.find_all('aside')[2].text.replace('\n', '').replace('\t', ''))
    services.append(', '.join(str(tag.text) for tag in soup.find_all('li', {'class': 'ie-company-service-item'})))
    adresses.append(soup.find('address').text.replace('\n', '').replace('\t', ''))
    
d = { 'nome': names, 'responsável': responsibles, 'especialidades': especialities, 'descrição': descriptions, 'serviços': services, 'endereço': adresses}
df = pd.DataFrame(data=d)
df = df[['nome', 'responsável', 'especialidades', 'serviços', 'endereço', 'descrição']]
df
```
