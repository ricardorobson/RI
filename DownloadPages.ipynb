```python
import requests as rq
import pandas as pd
import io
import os
import shutil
#pip install tldextract
import tldextract

from requests.packages.urllib3.exceptions import InsecureRequestWarning

rq.packages.urllib3.disable_warnings(InsecureRequestWarning)
```

```python
dataFrame = pd.read_csv('Links.csv')
dataFrame
```

### Download positive pages

```python
dir = 'Pages'
if os.path.exists(dir):
    shutil.rmtree(dir)
os.makedirs(dir)
os.makedirs(dir+'/Positives/')

index = 0
last_domain = ''

for url in dataFrame['P치ginas positivas']:
    page = rq.get(url, verify=False)
    domain = tldextract.extract(url).domain
    print(domain)
    
    if(domain != last_domain):
        print(last_domain)
        last_domain = domain
        os.makedirs(dir+'/Positives/'+domain)
        index = 0
    
    f = io.open(dir+'/Positives/'+domain+'/'+str(index)+'.html', 'w', encoding="utf-8")
    f.write(page.text)
    f.close()
    print(index)
    index += 1
```

### Download negative pages

```python
os.makedirs(dir+'/Negatives/')
index = 0
last_domain = ''

for url in dataFrame['P치ginas negativas'].dropna():
    page = rq.get(url, verify=False)
    domain = tldextract.extract(url).domain
    print(domain)
    
    if(domain != last_domain):
        index = 0
        print(last_domain)
        last_domain = domain
        if  os.path.exists(dir+'/Negatives/'+domain):
            index = len(os.listdir(dir+'/Negatives/'+domain))
        else:
            os.makedirs(dir+'/Negatives/'+domain)
    
    f = io.open(dir+'/Negatives/'+domain+'/'+str(index)+'.html', 'w', encoding="utf-8")
    f.write(page.text)
    f.close()
    print(index)
    index += 1
```
